```python
tfidf_matrix, _ = Embeddings.tfidf_vectorizer(df["document"])
tfidf_matrix
```

```python
svd_matrix, _ = Embeddings.svd(tfidf_matrix, 2)
svd_matrix
```

```python
pca_matrix, _ = Embeddings.pca(tfidf_matrix, 2)
pca_matrix
```

```python
lsa_matrix, _, _ = Embeddings.lsa(df["document"], n_components=2)
lsa_matrix
```
